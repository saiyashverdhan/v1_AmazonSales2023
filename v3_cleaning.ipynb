```python
import pandas as pd
import numpy as np
```

```python
# loading dataset from local storage
# using dtype as unicode which identifies entries in UTF-8 format, readable by pandas 
df = pd.read_csv("C:\\Job hunt\\databases\\Amazon full dataset\\Amazon-Products.csv",dtype='unicode')
```

```python
df = df.loc[:,['name','main_category','sub_category','discount_price','actual_price','ratings','no_of_ratings']]
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# finding percentage of missing data
for col in df.columns:
    percentage=(df[col].isnull().sum()/df.shape[0])*100
    print(col,round(percentage,2),"%")
```

```python
# Dropping entries with null actual price values
df.dropna(subset=['actual_price'],inplace=True)
df 
```

```python
# str.strip to remove leading and ending spaces
# using regex values to replace multiple string with the same string  
df_discount_cleaned = df.discount_price.str.strip().str.replace(r'[₹,]','',regex=True)
df_discount_cleaned.value_counts()
```

```python
pd.to_numeric(df_discount_cleaned)
```

```python
# performing same operations with actual_price column  
df_actual_cleaned = df.actual_price.str.strip().str.replace(r'[₹,]','',regex=True)
df_actual_cleaned.value_counts()
```

```python
pd.to_numeric(df_actual_cleaned)
```

```python
df['actual_price_clean'] = pd.to_numeric(df_actual_cleaned)
```

```python
df['discount_price_clean'] = pd.to_numeric(df_discount_cleaned)
```

```python
df.head(3)
```

```python
df_ratings = df.dropna(subset=['ratings'])
df_ratings.ratings.value_counts()
```

```python
# The errors='coerce' parameter returns any non-numeric values (like 'Get', '₹68.99', etc.) into NaN.
df_ratings['ratings'] = pd.to_numeric(df_ratings['ratings'],errors='coerce')
```

```python
# keeping only the rows where ratings are between 1 and 5
df_ratings = df_ratings[df_ratings['ratings'].between(1, 5.1)]
df_ratings
```

```python
df_ratings.to_excel("amazon_ratings.xlsx",sheet_name="cleaned_ratings")
```

```python
# dropping raw price columns in ratings datframe
df_ratings.drop(columns=['discount_price','actual_price'],axis=1)
```

```python
df_ratings.actual_price_clean.value_counts()
```

```python
# dropping raw price columns in main datframe
df = df.drop(columns=['discount_price','actual_price'],axis=1)
```

```python
# resetting the index and avoiding making a column out of the old index
df = df.reset_index(drop=True)
df
```

```python
df.to_excel("df_clean.xlsx", sheet_name="clean_main")
```

```python
# converting no_of_ratings into numeric values
df['no_of_ratings'] = pd.to_numeric(df.no_of_ratings.str.replace(",",""),errors="coerce")
```
